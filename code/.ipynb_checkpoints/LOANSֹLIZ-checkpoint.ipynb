## Pre-processing of the data 

```python
#look at 10 first rows
df.head(10)
#look at colums (name of variables)
df.columns
#size of data
df.shape

df.describe()
```

### Handle with missing values in the data 

```python
#Detect missing values in the data
#How many Nulls for each field?
null_columns=df.columns[df.isnull().any()]
df[null_columns].isnull().sum()
type(null_columns)
```

```python
#Percent of missing values for each field
print (df.isnull().mean())
#Keep only the fields where there is less than 25% of missing values
df = df.loc[:, df.isnull().mean() < .25]
#The remained fields
print(df.columns)
#df.head(10)
df.shape
#Delete rows with missing values
#percent of 'normal' values for each row
df.notnull().mean(axis=1)
#Keep only the rows where there are no missing values 
df = df.loc[df.notnull().mean(axis=1) == 1, :]
df.head(10)
df.shape

```

### Handle with variables containing only one value 

```python
#Delete fields with only one unique value
#Number of unique values for each field
df.nunique()
#Keep only the fields where there is more than one unique value
df = df.loc[:, df.nunique() > 1]
#The remained fields
print(df.columns)
df.head(10)
df.shape
```

### Encoding categorial variables to numeric variables

```python
#Converting problematic variables
#type of values in each column
for fld in df.columns : 
    print("type of field {0} is: {1}".format(fld, type(df[fld].iloc[0])))
```

```python
#Convert int_rate to a float and drop the % sign
type(df['int_rate'].iloc[0])
df['int_rate'] = df.int_rate.apply(lambda x: x.strip('%'))
df['int_rate'] = df.int_rate.apply(pd.to_numeric)
type(df['int_rate'].iloc[0])

#Convert revol_util to a float and drop the % sign
type(df['revol_util'].iloc[0])
df['revol_util'] = df.revol_util.apply(lambda x: x.strip('%'))
df['revol_util'] = df.revol_util.apply(pd.to_numeric)
type(df['revol_util'].iloc[0])

#Convert emp_length to a float and drop the the word years
df['emp_length'].value_counts()
df['emp_length'] = df.emp_length.apply(lambda x: x.strip('years+ <'))
df['emp_length'].value_counts()
df['emp_length'] = df.emp_length.apply(pd.to_numeric)
type(df['emp_length'].iloc[0])
```

```python
#Encoding of categorial variables to numerical
def coding(col, dct):
  colCoded = pd.Series(col, copy=True)
  for key, value in dct.items():
    colCoded.replace(key, value, inplace=True)
  return colCoded

df['loanStatusEnc'] = coding(df['loan_status'], {'Charged Off':0, 'Does not meet the credit policy. Status:Charged Off':0, 'Fully Paid':1, 'Does not meet the credit policy. Status:Fully Paid':1})
df['loanStatusEnc'].value_counts()

df['term'].unique()
df['termEnc'] = coding(df['term'], {' 36 months':0, ' 60 months':1})
df['termEnc'].value_counts()

df['home_ownership'].value_counts()
df['homeOwnEnc'] = coding(df['home_ownership'], {'RENT':0, 'MORTGAGE':1, 'OWN':2, 'OTHER':3})
df['homeOwnEnc'].value_counts()

df['grade'].value_counts()
df['gradeEnc'] = coding(df['grade'], {'A':0, 'B':1, 'C':2, 'D':3, 'E':4, 'F':5, 'G':6})
df['gradeEnc'].value_counts()
```
