```python
import pandas as pd
df = pd.read_csv("data_choice_withContextVal.csv")
```

```python
def a(datas,val,isPercent=False):
    count_z = 0
    for data in datas:

        if data == val:
            count_z += 1
    if isPercent:
        return 100*count_z/len(datas)
    return count_z
```

```python
df["NumErr"]=df.groupby(["Subject", "TrialType"])["Correct"].transform(lambda x: a(x.values,0))
df["NumErr%"]=df.groupby(["Subject", "TrialType"])["Correct"].transform(lambda x: a(x.values,0,True))
```
